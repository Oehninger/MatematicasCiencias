# Ejemplo 2: Modelo Lotka-Volterra

El sistema de ecuaciones diferenciales de la forma:
$$
    \begin{array}{rcl}
         \dot{x} &=& x(\beta -by),\\  \\
         \dot{y} &=& y(-\mu +cx).\\
    \end{array}
$$
donde $\beta$, $b$, $\mu$ y $c$ son constantes positivas.  Resolveremos numericamente esta ecuación usando odeint.

Implementamos esta ecuación en Python como sigue:

```python
import numpy as np
def grad(z, t, beta, b, mu, c):
    x, y = z
    dzdt = [x*(beta-b*y), y*(-mu+c*x)]
    return dzdt
```

Asignamos valores a los parámetros  $\beta$, $b$, $\mu$ y $c$:

```python
beta=0.7
b=0.3
mu=0.2
c=0.25
```

Consideramos una población inicial  $x(0)=3$ e $y(0)=1$.

```python
z0=[ 3, 1]
```

Generaremos una solución sobre el intervalo $[0,40]$ con una partición de 401 puntos. Lo que se define por:

```python
t=np.linspace(0,40,401)
```

Ya se han definido todos los aspectos necesarios para buscar una solución de la ecuación diferencial. Ahora debemos recurrir al comando ODEINT para generar la solución. Para que la función grad considere los parámetros $\beta$, $b$, $\mu$ y $c$, se los ingresamos usando args.

```python
from scipy.integrate import odeint
sol=odeint(grad, z0, t, args=(beta,b,mu,c))
```

La solución es un arreglo de tamaño (401,2). El siguiente código permite graficar la solución:

```python
import matplotlib.pyplot as plt
plt.plot(t, sol[:, 0], 'b', label='x(t)')
plt.plot(t, sol[:, 1], 'g', label='y(t)')
plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.show()
```
